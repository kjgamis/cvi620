```python
noise25 = np.random.normal(0, 25, img.shape).astype('uint8')
noise2 = np.random.normal(0, 2, img.shape).astype('uint8')
```

```python
noisy_img25 = cv2.add(img, noise25)
noisy_img2 = cv2.add(img, noise2)
```

```python
plt.figure(figsize=(25,25))
plt.subplot(131)
plt.imshow(img)

plt.subplot(132)
plt.imshow(noisy_img25)
plt.title("25")


plt.subplot(133)
plt.imshow(noisy_img2)
plt.title("2")


plt.show()
```

# Salt and Pepper Noise

```python
img = cv2.imread('../S3/Lucy.jpg')
```

```python
w, h, _ = img.shape
# w*h 50%
```

```python
number_of_pixels = np.random.randint(600, 10000)
for i in range(number_of_pixels):
    x = np.random.randint(0, w)
    y = np.random.randint(0, h)
    img[x,y] = 0

print(f"NOP: {number_of_pixels}")
for i in range(number_of_pixels):
    x = np.random.randint(0, w)
    y = np.random.randint(0, h)
    img[x,y] = 255
```

```python
cv2.imshow('frame', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
