# Matplotlib Tutorial

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
x = [i for i in range(1,7)]
y = [i**2 for i in x]

# y = x**2

plt.figure(figsize=(4,4))
plt.plot(x, y)
plt.title("math function: y=x**2")
plt.xlabel("x")
plt.ylabel('x**2')

# plt.grid(False)
plt.savefig('matplot diagram.png')
plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', markersize=16, linestyle='--', linewidth=4)
plt.show()
```

```python
x = [i for i in range(1, 7)]

y1 = x
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]


plt.plot(x, y1, label="y=x")
plt.plot(x, y2, label='y=x**2')
plt.plot(x, y3, label='y=x**3')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(131)
plt.plot(x, y1, label='x = y')
plt.title('x=y')
plt.legend()

plt.subplot(132)
plt.plot(x, y2, label='x = y**2')
plt.title('x=y**2')
plt.legend()

plt.subplot(133)
plt.plot(x, y3, label='y = x**3')

plt.legend()
plt.show()
```

```python
# Shallow copy vs. Deep copy
import copy

a = [4, 5, 6]
b = copy.deepcopy(a)
b = a.copy()

print(f'a: {a}')
print(f'b: {b}')


a[0] = 0

print(f'a: {a}')
print(f'b: {b}')
```

```python
img1 = cv2.imread("Lucy.jpg")
img2 = cv2.imread("Lucy.jpg")

print(type(img1), type(img2))

img1_copy = img1
img2_copy = img2.copy()

img1[0:100, 0:100] = 0
img2[100:200, 100:200] = 0


cv2.imshow('img1', img1)
cv2.imshow('img2', img2)
cv2.imshow('img2_copy', img2_copy)
cv2.imshow('img1_copy', img1_copy)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
img = cv2.imread('Lucy.jpg')
# print(img.shape)

rows, cols, _ = img.shape
max = 0 


for i in range(rows):
    for j in range(cols):
        if img[i, j, 2] > max:
            max = img[i, j, 2]


print(max)
```
