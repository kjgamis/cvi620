```python
cap = cv2.VideoCapture("S10_cars_video.mp4")

while True:
    ret, frame = cap.read()

    if frame is None:
        break

    cv2.imshow('frame', frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture("S10_cars_video.mp4")

desired_fps = 1

original_fps = int(cap.get(cv2.CAP_PROP_FPS))
interval = int(original_fps / desired_fps)
count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if count % interval == 0:
        cv2.imshow('frame', frame)

    if cv2.waitKey(1) == ord('q'):
        break

    count+=1

cap.release()
cv2.destroyAllWindows()
```
