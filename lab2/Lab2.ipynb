4. 
- Add padding to all four sides of the image using a border of 50 pixels.
Try at least three border types. What visual differences do you see?
- How can you change the color of padding

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image_orig = cv2.imread("dapper.jpg")
image_rgb = cv2.cvtColor(image_orig, cv2.COLOR_BGR2RGB)

# padding with different border types
# padding colour can be changed by changing the `value` arg in `cv2.copyMakeBorder`
border_constant = cv2.copyMakeBorder(image_rgb, 500, 500, 500, 500, cv2.BORDER_CONSTANT, value=[255, 0, 0])  # Red
border_reflect = cv2.copyMakeBorder(image_rgb, 500, 500, 500, 500, cv2.BORDER_REFLECT)
border_wrap = cv2.copyMakeBorder(image_rgb, 500, 500, 500, 500, cv2.BORDER_WRAP)

# Original image
plt.subplot(1, 4, 1)
plt.imshow(image_rgb)
plt.title("Original Image")

# Constant border
plt.subplot(1, 4, 2)
plt.imshow(border_constant)
plt.title("Constant Border")

# Reflect border
plt.subplot(1, 4, 3)
plt.imshow(border_reflect)
plt.title("Reflect Border")

# Wrap border
plt.subplot(1, 4, 4)
plt.imshow(border_wrap)
plt.title("Wrap Border")

plt.tight_layout()
plt.show()
```

6. Plot the function y = x^2+6 using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [i for i in range(1, 11)]
y = [i**2 + 6 for i in x]

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of function y")
plt.grid()
plt.show()
```

7. Plot sin(x) and cos(x) on the same figure. Add grid, labels, title, and legend.

```python
import matplotlib.pyplot as plt
from numpy import sin, cos
x = [i for i in range(1, 11)]
y1 = [i**2 + 6 for i in x]
y2 = sin(x)
y3 = cos(x)

plt.plot(x, y1, label="y = x^2 + 6")
plt.plot(x, y2, label="y = sin(x)")
plt.plot(x, y3, label="y = cos(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Plot of function y")
plt.grid()
plt.show()
```

8. Read an image and display it alongside its grayscaled version using matplotlib.

```python
import cv2
import matplotlib.pyplot as plt

image_orig = cv2.imread("dapper.jpg")
image_rgb = cv2.cvtColor(image_orig, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_orig, cv2.COLOR_BGR2GRAY)

# Original RGB image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original RGB")

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap="gray")
plt.title("Grayscale")

plt.tight_layout()
plt.show()
```

9. Having an image, create a 1x3 subplot that shows:

- Original RGB image
- Red channel
- HSV version
- Adjust the figure size for better visualization.

```python
import cv2
import matplotlib.pyplot as plt

image_orig = cv2.imread("dapper.jpg")
image_rgb = cv2.cvtColor(image_orig, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image_orig, cv2.COLOR_BGR2HSV)

# Original RGB image
plt.subplot(131)
plt.imshow(image_rgb)
plt.title("Original RGB")

# Red channel
plt.subplot(132)
plt.imshow(image_rgb[:, :, 0], cmap="Reds")
plt.title("Red Channel")

# HSV version
plt.subplot(133)
plt.imshow(image_hsv)
plt.title("HSV Version")

plt.tight_layout()
plt.show()
```

10. Search about Histograms and explain them. Plot a sample histogram below.

```python
# A histogram is a graph type that represents the distribution of numerical data by showing the number of data points that fall within a specified range.
# It represents the frequency of occurrence of different values in a dataset.

import matplotlib.pyplot as plt
import numpy as np

# Generate random scores out ot 100
data = np.random.randint(0, 100, size=1000)

plt.hist(data, bins=30, color='blue', edgecolor='black')
plt.title("Score Histogram")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
